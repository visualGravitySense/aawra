```python
def GetComments(username):
    user_id = bot.get_user_id_from_username(username)
    medias = bot.get_user_medias(user_id,filtration=False)
    comments = [[i['text'] for i in bot.get_media_comments(m)] for m in medias]
    comments = [x for y in comments for x in y]
    return comments
```

```python
def GetLikers(username):
    user_id = bot.get_user_id_from_username(username)
    medias = bot.get_user_medias(user_id,filtration=False)
    likers = [bot.get_media_likers(i) for i in medias]
    likers = list(set([x for y in likers for x in y]))
    return likers
```

```python
def ShowLikersInfo(username, n):
    likers = GetLikers(username)
    print(f'Likers: {len(likers)}')
    userinfos = [bot.get_user_info(str(i)) for i in likers[:n]]
    return pd.DataFrame([[u['username'], u['follower_count'], u['following_count'], u['full_name'], u['biography']] for u in userinfos], columns=['username','follower_count','following_count','full_name','biography'])
```
